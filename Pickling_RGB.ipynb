```python
import os
import cv2
import numpy as np
import pickle
```

```python
img_folder = os.path.join(os.getcwd(),'rawdata')

#connects the path to the required folder, where images are to be joined
```

```python
os.getcwd()

#gets the current directory, where the path is to be located
```

```python
images = []
labels = []
```

```python
for i in os.listdir(img_folder):
    print(i)
    
#list the number of captured directories/folders
```

```python
for i in os.listdir(img_folder):
    for j in os.listdir(os.path.join(img_folder,i)):    
        print(j)
        
        
#goes to per folder and captures each folder, i.e captures the content inside the directory
```

```python
for i in os.listdir(img_folder):
    for j in os.listdir(os.path.join(img_folder,i)):
        image_path = os.path.join(img_folder,i,j)
        image = cv2.imread(image_path)
        if image is None:
            print('Wrong path:', image_path)
        else:
            img = cv2.resize(image, dsize=(128,128))
            images.append(img)
        #images.append(image)
            labels.append(i)
    print(i,'------> completed')
    
#storing all the images from all the directories to 'images'
#storing all the labels in correspondence to the 'images' in 'labels' 
```

```python
len(labels)

#number to labels
```

```python
set(labels)

#unique entries in labels
```

```python
images = np.array(images)
labels = np.array(labels)

#converting the 'images' and 'labels' to np arrays for pickling
```

```python
with open(os.path.join(img_folder,'images.p'),'wb') as f:
    pickle.dump(images,f)
#pickling all the images to one file 'images.p'
#'wb' is for write mode, for writing all the images to the pickle file

with open(os.path.join(img_folder,'labels.p'),'wb') as f:
    pickle.dump(labels,f)
#pickling all the labels to 1 file 'labels.p'

#the command opens a file for writing binary data
#it constructs the file path using os.path.join()
#and then uses pickle.dump() to save the images object to the file. 
#This is commonly done to store data in a serialized format for later retrieval or to transfer it between different programs or environments. 
#The file extension .p suggests that the data is being saved in a pickle format.
```
